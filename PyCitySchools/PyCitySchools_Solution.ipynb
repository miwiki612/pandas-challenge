# PyCity Schools Analysis

### Definition

- Charter school is a public school that operates as a school of choice.
- District school is a public school under the control of a local/district school board.

According to the definition, we could notice a Charter school is an additional option and a District school is a fundamental education supplier. How those two types of school are funded is cohesive with the performance.

### District Summary

In this study, 15 schools and a total 39,170 students are included. Overall, the average score on Math is 78.99, Reading is 81.88. When a score is higher than or equal to 70 as pass, 74.88% of students pass Math, 85.81% of students pass Reading and 65.17% students pass both Math and Reading. In general Math could be more challenging for students and for individual students their performance on Math and Reading could be very different.

### School Summary

- Highest-Performing Schools

The top 5 schools are schools with 97.04% to 95.95% of students passing both Math and Reading. All of the top schools are Charter schools. Passing rate on Math is about 2.9% less than Reading.

Cabrera High School has the highest overall passing rate, 97.04%.

- Bottom Performing Schools

The bottom 5 schools are schools with 53.54% to 52.99% of students passing both Math and Reading. All of the bottom schools are District schools. Passing rate on Math and Reading have large variance, Math is about 14.7% less than Reading. The performance on Math is significantly worse compared to Reading.

Rodriguez High School has the lowest overall passing rate, 52.99%.

- Math and Reading Scores by Grade

Average score has no significant difference between grades for individual schools. Standard deviation on score is less than 1 for both Math and Reading in all grades. However the performance for schools are variance. Average Math score’s standard deviation is 3.40 and Reading score’s standard deviation is 1.57.

Compare average score and passing rate, it’s obvious to point out the students in better performing schools have generally similar good performance. On the other hand, in a school without ideal performance, students’ performance can be very good or very bad. Therefore schools have similar average scores and yet the passing rate is significantly different.

### Scores by School Spending, School Size and School Type

In general, schools with lower Spending Ranges (Per Student), less than $585 and smaller size, less than 2000 students have better performance. School has higher Spending Ranges (Per Student) and larger size has less ideal performance.

Additionally,  Spending Ranges (Per Student) and School Size are both correlated to School Type. Charter schools usually have less Spending Ranges (Per Student) and smaller in size. District school is opposite.

- Summary

Charter schools have better performance than District schools. Since by definition Charter school is an optional school, they won’t exist in the first place if the performance is not significantly better.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len( school_data_complete[ 'School ID' ].value_counts() ) # potential same name, count by ID
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete[ 'Student ID' ].nunique() # some students have same name, count by ID
student_count
```

```python
# Calculate the total budget
# test if all schools are included in complete data frame. if so use school_data to avoid duplicate.
if school_data[ 'School ID' ].nunique() == school_data_complete[ 'School ID' ].nunique() : print("All school are included in complete analysis")
if school_data[ 'School ID' ].nunique() == len(school_data['School ID'] ) : print("No duplicate entries")
total_budget = school_data[ 'budget' ].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete[ 'math_score' ].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete[ 'reading_score' ].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (reading scores greather than or equal to 70)
passing_reading_count = school_data_complete[ (school_data_complete[ 'reading_score' ] >= 70 )].count()[ 'student_name' ]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary= pd.DataFrame( {
    "Total Schools" : [school_count] , "Total Students": [student_count] , "Total Budget" : [total_budget],
    "Average Math Score" : [average_math_score] , "Average Reading Score" : [average_reading_score] ,
    "% Passing Math" : [passing_math_percentage] , "% Passing Reading" : [passing_reading_percentage] ,
    "% Overall Passing" : [overall_passing_rate]
} )

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary['Average Math Score'] = district_summary['Average Math Score'].map("{:,.2f}".format)
district_summary['Average Reading Score'] = district_summary['Average Reading Score'].map("{:,.2f}".format)
district_summary['% Passing Math'] = district_summary['% Passing Math'].map("{:,.2f}%".format)
district_summary['% Passing Reading'] = district_summary['% Passing Reading'].map("{:,.2f}%".format)
district_summary['% Overall Passing'] = district_summary['% Overall Passing'].map("{:,.2f}%".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_complete['type'].unique()
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby(['school_name', 'type']).count()['Student ID'].to_frame( name="Total Students")

# prepare type column for merage later
per_school_counts = per_school_counts.reset_index()
per_school_counts = per_school_counts.set_index("school_name")
per_school_counts = per_school_counts.rename( columns={"type":"School Type"} )

per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data.groupby(['school_name']).mean(numeric_only=True)['budget'].to_frame( name="Total School Budget")
per_school_capita = (per_school_budget['Total School Budget'] / per_school_counts['Total Students'] ).to_frame( name="Per Student Budget")
```

```python
# Calculate the average test scores per school
per_school_math = school_data_complete.groupby(['school_name']).mean(numeric_only=True)['math_score'].to_frame( name = "Average Math Score")
per_school_reading = school_data_complete.groupby(['school_name']).mean(numeric_only=True)['reading_score'].to_frame( name = "Average Reading Score")
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[ (school_data_complete["math_score"] >= 70) ]
school_students_passing_math = students_passing_math.groupby(['school_name']).size()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[ (school_data_complete["reading_score"] >= 70) ]
school_students_passing_reading = students_passing_reading.groupby(['school_name']).size()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
```

```python
# Use the provided code to calculate the passing rates
# add to prepare for merage
per_school_passing_math = (school_students_passing_math / per_school_counts['Total Students'] * 100).to_frame( name="% Passing Math" )
per_school_passing_reading = (school_students_passing_reading / per_school_counts['Total Students'] * 100 ).to_frame( name="% Passing Reading" )
overall_passing_rate = (school_students_passing_math_and_reading / per_school_counts['Total Students'] * 100 ).to_frame( name="% Overall Passing")
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = pd.merge( per_school_counts , per_school_budget , how="left", on="school_name" )
per_school_summary = pd.merge( per_school_summary , per_school_capita , how="left",  on="school_name" )
per_school_summary = pd.merge( per_school_summary , per_school_math , how="left",  on="school_name" )
per_school_summary = pd.merge( per_school_summary , per_school_reading , how="left",  on="school_name" )
per_school_summary = pd.merge( per_school_summary , per_school_passing_math , how="left",  on="school_name" )
per_school_summary = pd.merge( per_school_summary , per_school_passing_reading , how="left",  on="school_name" )
per_school_summary = pd.merge( per_school_summary , overall_passing_rate , how="left",  on="school_name" )

per_school_summary.index.name = None

# backup numeric data frame before formatting
per_school_summary_num = per_school_summary.copy()
per_school_summary_num.dtypes
```

```python
# Formatting
# add $
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
# reduce decimal
per_school_summary['Average Math Score'] = per_school_summary['Average Math Score'].map("{:,.2f}".format)
per_school_summary['Average Reading Score'] = per_school_summary['Average Reading Score'].map("{:,.2f}".format)
# add %
per_school_summary['% Passing Math'] = per_school_summary['% Passing Math'].map("{:,.2f}%".format)
per_school_summary['% Passing Reading'] = per_school_summary['% Passing Reading'].map("{:,.2f}%".format)
per_school_summary['% Overall Passing'] = per_school_summary['% Overall Passing'].map("{:,.2f}%".format)

# Display the DataFrame

per_school_summary
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]
```

```python
# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby(['school_name']).mean(numeric_only=True)['math_score'].to_frame( name="9th" )
tenth_grader_math_scores = tenth_graders.groupby(['school_name']).mean(numeric_only=True)['math_score'].to_frame( name="10th" )
eleventh_grader_math_scores = eleventh_graders.groupby(['school_name']).mean(numeric_only=True)['math_score'].to_frame( name="11th" )
twelfth_grader_math_scores = twelfth_graders.groupby(['school_name']).mean(numeric_only=True)['math_score'].to_frame( name="12th" )

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.merge( ninth_grade_math_scores , tenth_grader_math_scores , on="school_name", how="outer" )
math_scores_by_grade = pd.merge( math_scores_by_grade , eleventh_grader_math_scores , on="school_name", how="outer" )
math_scores_by_grade = pd.merge( math_scores_by_grade , twelfth_grader_math_scores , on="school_name", how="outer" )

# Minor data wrangling
math_scores_by_grade.index.name = None

# not required in original start - stdard deviation of grades in each school.
math_scores_by_grade["SD"] = math_scores_by_grade.std(axis=1)

# Display the DataFrame
math_scores_by_grade
```

```python
# not required in original start - stdard deviation for schools.
math_scores_by_grade.std(axis=0).head(4).mean()
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby(['school_name']).mean(numeric_only=True)['reading_score'].to_frame( name="9th" )
tenth_grader_reading_scores = tenth_graders.groupby(['school_name']).mean(numeric_only=True)['reading_score'].to_frame( name="10th" )
eleventh_grader_reading_scores = eleventh_graders.groupby(['school_name']).mean(numeric_only=True)['reading_score'].to_frame( name="11th" )
twelfth_grader_reading_scores = twelfth_graders.groupby(['school_name']).mean(numeric_only=True)['reading_score'].to_frame( name="12th" )

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.merge( ninth_grade_reading_scores , tenth_grader_reading_scores , on="school_name", how="outer" )
reading_scores_by_grade = pd.merge( reading_scores_by_grade , eleventh_grader_reading_scores , on="school_name", how="outer" )
reading_scores_by_grade = pd.merge( reading_scores_by_grade , twelfth_grader_reading_scores , on="school_name", how="outer" )


# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# not required in original start - stdard deviation of grades in each school.
reading_scores_by_grade["SD"] = reading_scores_by_grade.std(axis=1)

# Display the DataFrame
reading_scores_by_grade
```

```python
# not required in original start - stdard deviation for schools.
reading_scores_by_grade.std(axis=0).head(4).mean()
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
# use numeric copy for calculation
school_spending_df = per_school_summary_num.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut( school_spending_df['Per Student Budget'] , bins = spending_bins , labels = labels )

# Formatting
# add $
school_spending_df["Total School Budget"] = school_spending_df["Total School Budget"].map("${:,.2f}".format)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame

school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({ 'Average Math Score' : spending_math_scores , 'Average Reading Score' : spending_reading_scores ,
                                '% Passing Math' : spending_passing_math, '% Passing Reading' : spending_passing_reading,
                                '% Overall Passing' : overall_passing_spending, },
                                index = spending_math_scores.index )

# Display results
spending_summary
```

```python
# not required in original start - find school type and spending group
school_type_spending_df = school_spending_df.reset_index().groupby(["Spending Ranges (Per Student)", "School Type" ]).count()["index"].to_frame(name = "School Count")

school_type_spending_df
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# use numeric copy for calculation
per_school_summary = per_school_summary_num.copy()

# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
per_school_summary["School Size"] = pd.cut( per_school_summary['Total Students'] ,bins = size_bins, labels=labels )

# Formatting
# add $
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame( {"Average Math Score" : size_math_scores, "Average Reading Score" : size_reading_scores,
                             "% Passing Math" : size_passing_math, "% Passing Reading" : size_passing_reading , 
                             "% Overall Passing" : size_overall_passing } ,
                           index = size_math_scores.index )

# Display results
size_summary
```

```python
# not required in original start - find school size and spending
school_type_size_df = per_school_summary.reset_index().groupby(["School Size", "School Type" ]).count()["index"].to_frame(name = "School Count")

school_type_size_df
```
